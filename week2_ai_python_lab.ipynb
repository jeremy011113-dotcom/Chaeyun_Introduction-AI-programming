## 1. 제어문 (Control Flow)

### 1.1. if, elif, else
조건에 따라 다른 코드를 실행할 때 사용합니다.

```python
score = 85

if score >= 90:
    print("A 학점입니다.")
elif score >= 80:
    print("B 학점입니다.")
elif score >= 70:
    print("C 학점입니다.")
else:
    print("재수강이 필요합니다.")
```

### 1.2. for 반복문
리스트, 튜플, 문자열 등 순회 가능한(iterable) 객체의 요소를 차례대로 꺼내어 반복 실행합니다.

```python
fruits = ["사과", "바나나", "체리"]

for fruit in fruits:
    print(f"과일: {fruit}")
```

### 1.3. while 반복문
주어진 조건이 참(True)인 동안 계속해서 블록의 코드를 실행합니다.

```python
count = 5
while count > 0:
    print(f"카운트 다운: {count}")
    count -= 1
print("발사!")
```

## 2. 함수와 모듈 (Functions & Modules)

### 2.1. 함수 정의와 호출
반복적으로 사용되는 코드를 하나로 묶어 재사용 가능하게 만드는 기능입니다.

```python
def calculate_average(scores):
    """리스트로 주어진 점수들의 평균을 계산하는 함수"""
    if not scores:
        return 0
    return sum(scores) / len(scores)

my_scores = [80, 90, 100, 75]
avg = calculate_average(my_scores)
print(f"내 점수의 평균은 {avg}점 입니다.")
```

## 3. 객체 지향 프로그래밍 기초 (OOP)

### 3.1. 클래스(Class)와 객체(Object)
상태(속성, Attributes)와 행위(메서드, Methods)를 하나로 묶은 데이터 모델을 설계도(클래스)로 정의하고,
이를 통해 실체(객체)를 생성합니다.

```python
class Robot:
    # 생성자 (초기화 메서드)
    def __init__(self, name, model, vendor):
        self.vendor = vendor
        self.name = name
        self.model = model
        self.battery = 100
        
    def greet(self):
        print(f"안녕하세요! 저는 {self.vendor}에서 만든 {self.model} 모델 {self.name} 입니다.")
        
    def work(self):
        if self.battery >= 10:
            print(f"{self.name}가 작업을 수행합니다. 삐빅.")
            self.battery -= 10
        else:
            print("배터리가 부족합니다. 충전해주세요.")

# 클래스를 기반으로 두 개의 서로 다른 객체 생성
robot_a = Robot(name="알파", model="atlas", vendor="BostonDynamics")
robot_b = Robot(name="오메가", model="optimus", vendor="Tesla")

# 각 객체의 작동
robot_a.greet()
robot_a.work()
robot_b.work()
print(f"{robot_a.name}의 현재 배터리: {robot_a.battery}%")
```

## 4. [심화] 절차적 프로그래밍 vs 객체지향 프로그래밍
같은 목적의 프로그램을 절차적 방식과 객체지향 방식으로 각각 짜보고 구조의 차이를 비교해봅시다.

```python
# [강의용 코드] 절차적 프로그래밍 구현 예시 (은행 로직)
account_balance = 0

def deposit(amount):
    global account_balance
    account_balance += amount
    print(f"[절차적] {amount}원 입금 완료. 잔액: {account_balance}원")

deposit(1000)
deposit(500)
```

```python
# [강의용 코드] 객체지향 프로그래밍 구현 예시 (은행 객체)
class BankAccount:
    def __init__(self, owner):
        self.owner = owner
        self.balance = 0
    
    def deposit(self, amount):
        self.balance += amount
        print(f"[{self.owner}의 은행객체] {amount}원 입금 완료. 잔액: {self.balance}원")

my_account = BankAccount("학생A")
my_account.deposit(1000)
my_account.deposit(500)
```

```python
# TODO: [실습] 
# 여러분만의 '게임 캐릭터'를 절차적(변수+함수) 방식과 객체지향(클래스) 방식으로 각각 만들어보고 어떻게 다른지 느껴보세요.
# [1. 절차적 방식: 캐릭터 2명]

# 1) 두 캐릭터의 변수를 각각 따로 생성 관리
hero1_name = "용사1"
hero1_stamina = 10

hero2_name = "용사2"  # '용사2' 변수 추가
hero2_stamina = 10    # '용사2' 스태미너 추가

def move_forward(name, stamina):
    if stamina > 0:
        stamina -= 1
        print(f"{name}가 앞으로 1칸 이동했습니다! (남은 스태미너: {stamina}/10)")
    else:
        print(f"{name}는 지쳐서 움직일 수 없습니다...")
    return stamina

print("=== 절차적 방식: 2명 이동 ===")

# 누가 움직이는지 변수를 헷갈리지 않게 정확히 넣어주는 것이 중요해집니다.
hero1_stamina = move_forward(hero1_name, hero1_stamina)  # 용사1 이동
hero2_stamina = move_forward(hero2_name, hero2_stamina)  # 용사2 이동
hero2_stamina = move_forward(hero2_name, hero2_stamina)  # 용사2 한 번 더 이동!
```

```python
# [2. 객체지향 방식: 캐릭터 2명]

# 1) 클래스 정의 (설계도는 캐릭터가 몇 명이든 한 번만 만들면 됨)
class Character:
    def __init__(self, name):
        self.name = name     
        self.stamina = 10    
        
    def move_forward(self):
        if self.stamina > 0:
            self.stamina -= 1
            print(f"{self.name}가 앞으로 1칸 이동했습니다! (남은 스태미너: {self.stamina}/10)")
        else:
            print(f"{self.name}는 지쳐서 움직일 수 없습니다...")

print("\n=== 객체지향 방식: 2명 이동 ===")

# 2) 설계도로 객체 2개 뚝딱 생성! 변수를 따로 파고 관리할 필요 없음
hero1 = Character("용사1")
hero2 = Character("용사2")

# 3) 각자에게 명령만 내리면, 자기 스태미너는 자기가 알아서 깎음
hero1.move_forward()  # 용사1 이동
hero2.move_forward()  # 용사2 이동
hero2.move_forward()  # 용사2 한 번 더 이동!
```

## 5. [심화] 지연 평가 (Lazy Evaluation) 및 제너레이터
메모리를 효율적으로 사용하기 위해, 즉시 계산하지 않고 값이 실제로 필요하여 호출될 때만 계산하는 방식입니다.

```python
# [강의용 코드] 무한한 크기의 리스트를 만들고 싶다면? -> 제너레이터(Generator) 사용 (지연 평가)
def infinite_number_generator():
    num = 1
    while True:
        yield num  # 값을 기억해두고 함수를 '일시정지' 시킵니다. (Lazy의 핵심!)  양보하는 느낌. 다시 호출(next 등)하기 전까지는 계속 중단된 상태 유지.
        num += 1

gen = infinite_number_generator()
print("연산 시작!")
print("처음 하나:", next(gen))
print("그 다음 하나:", next(gen))
print("또 다음 하나:", next(gen))
```

## 6. [심화] 자동 미분 (Automatic Differentiation) 체험
JAX 같은 프레임워크가 복잡한 수식의 기울기를 자동 미분(Auto-grad)으로 어떻게 명시적으로 구하는지 살펴봅니다.

```python
# [강의용 코드] JAX를 활용한 명시적인 자동미분 체험
try:
    import jax
    import jax.numpy as jnp
    
    # 1. 미분할 함수 정의: f(x) = x^2
    def f(x):
        return x ** 2
    
    # 2. 미분(Gradient)을 계산해주는 도함수를 생성합니다. (명시적 자동미분)
    df_dx = jax.grad(f)
    
    x_val = 3.0
    print(f"원래 함수 연산값 f({x_val}):", f(x_val))
    
    # 3. 도함수에 값을 넣어서 미분값을 구합니다. (dy/dx = 2x 이므로 4.0이 나와야 합니다)
    print(f"x={x_val} 에서의 기울기(미분값):", df_dx(x_val))
except ImportError:
    print("HINT: jax가 설치되어 있지 않습니다. '!pip install jax jaxlib'를 실행해주세요.")
```

## 7. [번외] 지연 평가와 자동 미분의 놀라운 관계
우리가 배운 5번(Lazy Evaluation)과 6번(Auto Differentiation) 개념은 현대 AI 프레임워크(JAX, TensorFlow 등)에서 **불가분의 관계**에 있습니다.

1. **연산 그래프(Graph) 생성**: 컴퓨터가 미분(연쇄 법칙)을 거꾸로 수행하려면 수식이 어떤 순서로 합쳐졌는지 구조를 기록해두어야 합니다. 코드가 텍스트로 주어졌을 때 곧바로 값을 '즉시 평가(Eager Evaluation)' 해버리면 기록이 남지 않습니다. 따라서 일단 연산 과정만 기억해두는 **지연 평가(Lazy)** 철학이 적용됩니다.

2. **JAX 코어 원리**: JAX에서 파이썬 코드를 실행할 때 매번 진짜 연산을 돌리지 않고, 내부적으로 연산 추적(Tracing)만 수행하여 거대한 연산 그래프를 설계합니다. 수식이 끝난 후 오차를 역전파하기 위해 `jax.grad()`를 호출하면 컴파일 타임에 모든 것이 지연 평가 방식을 통해 최고 속도로 계산됩니다.
